# Урок 1. Введение в задачу классификации. Постановка задачи и подготовка данных.

1. **Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.**

Примеры случаев, когда важно максимизировать Precision:

*   Модель отбора по фото наличия у животных характерной особенности для селектирования и выведения новых пород.
*   Модель предсказания наличия у групп людей определенных способностей, например, наличия у человека способностей эффективного управления людьми при приеме на работу на ответственную руководящую должность.

Примеры случаев, когда важно максимизировать Recall:

*   Модель предсказания погоды, например, на море или на Крайнем севере, чтобы избежать кораблекрушения во время шторма, или пропадания людей во время снежной бури.
*   Модель контроля качества на производстве, чтобы не допустить попадания брака в готовую продукцию.




2. **Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?**

Потому, что F-мера позволяет находить наибольшее лучшее соотношение обеих противостоящих в качестве модели метрик Precision и Recall. Изменяя коэффициент β, можно усиливать одну и ослаблять другую метрику:

F = (β^2+1)(Precision × Recall) / (β^2Precision + Recall),

где β:
*   0 < β < 1 - выше Precision, ниже Recall;
*   β > 1 - ниже Precision, выше Recall;
*   β = 1 - сбалансированная F-мера.
