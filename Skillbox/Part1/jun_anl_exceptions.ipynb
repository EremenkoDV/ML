### Вопросы по лекциям.

Как поймать вообще все ошибки, которые могут произойти?

**Ответ:**

Для этого нужно заключить код программы в блок try..except, и в конструкции except "ловить" общий класс исключений Exception. Также возможна инструкция except без аргументов, что одно и тоже. Однако, "ловить" все исключения сразу не рекомендуется. Сначала следует отлавливать возможные частные случаи исключений типа ValueError или ZeroDivisionError, а в конце уже "ловить" оставшиеся.


Сколько раз подряд можно указывать except?

**Ответ:**

Неограниченно.

Вы хотите с помощью print вывести название ошибки в консоль, как это сделать?

**Ответ:**


```python
try:
    print(3/0)
except Exception as e:
    print(type(e).__name__)
```

Вы хотите с помощью print вывести параметры ошибки в консоль, как это сделать?

**Ответ:**


```python
try:
    print('3'/0)
except Exception as e:
    for arg in e.args:
        print(arg)
```

Что такое DeprecationWarning?

**Ответ:**

Это категория предупреждений о функциональности нежелательной к использованию. Эту категорию обычно используют для указания на то, что некая часть функциональности морально устарела (возможно ей на смену пришла более совершенная) и не рекомендуется к использованию.

### Задание 1.

Есть файл с протоколом регистраций пользователей на сайте (registrations.txt).<br>
Каждая строка содержит информацию о имени, электронной почте и возрасте человека. <br><br>


Надо проверить данные из файла, для каждой строки:
 - присутсвуют все три поля
 - поле имени содержит только буквы
 - поле email содержит @ и .
 - поле возраст является числом от 10 до 99<br>

В результате проверки нужно сформировать два файла
 - registrations_good.log для правильных данных, записывать строки как есть
- registrations_bad.log для ошибочных, записывать строку и вид ошибки.<br>

Для валидации строки данных написать метод, который может выкидывать исключения:
 - НЕ присутсвуют все три поля: ValueError
 - поле имени содержит НЕ только буквы: NotNameError (кастомное исключение)
 - поле email НЕ содержит @ и .(точку): NotEmailError (кастомное исключение)
 - поле возраст НЕ является числом от 10 до 99: ValueError
Вызов метода обернуть в try-except.

```python
import os.path
import re

filename = 'registrations_.txt'
filename_good = 'registrations_good.log'
filename_bad = 'registrations_bad.log'
```

```python
class NotNameError(ValueError):
	"""Exception raised for errors if
    the field of name contains not only letters.
	Исключение возбуждается для ошибок если
    поле имени содержит НЕ только буквы.
 
	Attributes:
		expression -- input expression in which the error occurred
				   -- входное выражение, в котором произошла ошибка
		message -- explanation of the error
				-- объяснение ошибки
	"""
	def __init__(self, value, message):
		self.value = value
		self.message = message


class NotEmailError(ValueError):
	"""Exception raised for errors if
    the field of email not contains @ и dot.
	Исключение возбуждается для ошибок если
    поле email НЕ содержит @ и точку.
 
	Attributes:
		expression -- input expression in which the error occurred
				   -- входное выражение, в котором произошла ошибка
		message -- explanation of the error
				-- объяснение ошибки
	"""
	def __init__(self, value, message):
		self.value = value
		self.message = message


def validation(line):
    fields = line.split('\n')[0].split(' ')
    if len(fields) != 3:
        raise ValueError(line, 'НЕ присутствуют все три поля.')
    if re.match('^[a-zA-Zа-яА-ЯёЁ]+$', fields[0]) is None:
        raise NotNameError(line, 'Поле #1 [Имя] содержит НЕ только буквы.')
    if re.match('(\w)+@(\w)+\.', fields[1]) is None:
        raise NotEmailError(line, 'Поле #2 [Email] НЕ содержит @ и точку.')
    if re.match('^[1-9][0-9]$', fields[2]) is None:
        raise ValueError(line, 'Поле #3 [Возраст] НЕ является числом от 10 до 99.')
    return
```

```python
total_good = 0
total_bad = 0

if os.path.exists(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        try:
            fg = open(filename_good, 'w', encoding='utf-8')
            fb = open(filename_bad, 'w', encoding='utf-8')
        except Exception as e:
            print(f'{e}')
        for line in f:
            try:
                line = line.split('\n')[0]
                validation(line)
            except NotNameError as e:
                print(f'{e}')
                fb.write(f'Строка: "{line}", ошибка: "{e}"...\n')
                total_bad += 1
            except NotEmailError as e:
                print(f'{e}')
                fb.write(f'Строка: "{line}", ошибка: "{e}"...\n')
                total_bad += 1
            except ValueError as e:
                print(f'{e}')
                fb.write(f'Строка: "{line}", ошибка: "{e}"...\n')
                total_bad += 1
            except Exception as e:
                print(f'{e}')
                fb.write(f'Строка: "{line}", ошибка: "{e}"...\n')
                total_bad += 1
            except RuntimeError as e:
                print(f'Неизвестная ошибка в строке "{line}": "{e}""...')
                fb.write(f'Строка: "{line}", ошибка: "{e}"...\n')
                total_bad += 1
            else:
                fg.write(f'{line}\n')
                total_good += 1
                
        print(f'Всего: хороших строк: {total_good}, плохих строк: {total_bad}.')
        fg.close
        fb.close    
        f.close
else:
    print(f'Файл {filename} не найден.')
```
